```python
from transformers import M2M100ForConditionalGeneration
from tokenization_small100 import SMALL100Tokenizer
```

tokenization_small100 is a python file which is accompied along with the `alirezamsh/small100` model but is needed to be downloaded seperately 

```python
model = M2M100ForConditionalGeneration.from_pretrained("alirezamsh/small100")
tokenizer = SMALL100Tokenizer.from_pretrained("alirezamsh/small100")
```

```python
text = "I am Achintya and I live in Bengaluru"
```

```python
tokenizer.tgt_lang = "de"
encoded_de = tokenizer(text, return_tensors="pt")
generated_tokens = model.generate(**encoded_de)
tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
```

```python
tokenizer.tgt_lang = "hi"
encoded_hi = tokenizer(text, return_tensors="pt")
generated_tokens = model.generate(**encoded_hi)
tokenizer.batch_decode(generated_tokens, skip_special_tokens=False)
```

```python
tokenizer.tgt_lang = "ja"
encoded_ja = tokenizer(text, return_tensors="pt")
generated_tokens = model.generate(**encoded_ja)
tokenizer.batch_decode(generated_tokens, skip_special_tokens=False)
```

The output given is I am Akinthya and I live in Bengaluru.

Model's Accuracy and Handling of Proper Nouns

The translations demonstrate the model's ability to translate the given sentence into different languages. However, an interesting observation is made regarding the translation of the name "Achintya" into Hindi: "the name was not translated into Hindi".

This behavior is common in machine translation, especially with proper nouns like names and specific place names (like "Bengaluru" which also remains untranslated in the German output). Translation models are typically trained to translate meaning rather than transliterate or translate proper nouns.

There are several reasons why "Achintya" was not translated into Hindi:

Proper Nouns are Often Transliterated, Not Translated: Names generally don't have direct translations across languages. Instead, they are often transliterated, meaning they are converted into the closest phonetic equivalent using the target language's script. In this case, "Achintya" is kept as "Achintya" in the Devanagari script for Hindi, which is the correct approach for a proper noun.
Out-of-Vocabulary (OOV) Words: If a name is very rare or not present in the model's training data, the model might not have learned a specific transliteration or translation for it.
Domain Specificity: General-purpose translation models might not have specific rules or extensive training data for all possible names.
Model Design: The M2M100 model is designed for many-to-many multilingual translation. Its objective is to capture the semantic meaning of sentences, and for proper nouns, preserving their identity through transliteration is often preferred over attempting a potentially incorrect translation.

The translations for "I am Achintya and I live in Bengaluru" across German, Hindi, and Japanese appear to be grammatically correct and convey the original meaning, with the proper nouns handled by transliteration rather than translation, which is standard practice.
